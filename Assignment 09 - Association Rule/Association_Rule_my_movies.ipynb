```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv(r"C:\Users\prajw\Data Science\Assignments\9 Association Rule\my_movies.csv")
```

```python
df=data.copy()
df.head()
```

```python
df=df.iloc[:,5:]
df
```

```python
frequent_items= apriori(df,min_support=0.1,use_colnames= True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric="lift",min_threshold=2.0)
rules
```

```python
rules.sort_values("lift",ascending = False)[0:20]
```

```python
rules[["support","confidence"]].hist()
plt.show()
```

```python
rules[["support","confidence","lift"]].hist(figsize=(15,7))
plt.show()
```

```python
plt.scatter(rules["support"],rules["confidence"])
plt.show()
```
